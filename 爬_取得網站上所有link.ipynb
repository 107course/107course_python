```python
import requests
from bs4 import BeautifulSoup
res = requests.get("http://apps.nknush.kh.edu.tw/ZeroBB")
soup = BeautifulSoup(res.text,'lxml')
for link in soup.select('a'):
    print(link.attrs)
    print(link.attrs['href'] +" : "+link.text)

#獲取所有 link
```
